Imported required libraries

```python
df = pd.read_csv('sample_data/Q1_Dataset.csv') #reading the provided dataset
df = df.drop_duplicates()
duplicate_rows = df[df.duplicated()]
print("Duplicate rows found:", len(duplicate_rows))
```

```python
#selecting numerical columns
num_cols = df.select_dtypes(include=['number'])
```

```python
#getting summary of numerical columns (min,max,standard deviation, quartiles)
summary = num_cols.describe().T
```

```python
#adding the median
summary['median']=num_cols.median()
```

```python
#renaming the columns to show the required data
summary =summary.rename(columns={
    'min': 'Minimum',
    'max': 'Maximum',
    'mean': 'Mean',
    'std': 'Standard Deviation',
    '25%': 'First Quartile',
    '50%': 'Median',
    '75%': 'Third Quartile'
})
```

```python
#selecting the required columns
summary_tab = summary[[
    'Minimum',
    'Standard Deviation',
    'First Quartile',
    'Median',
    'Third Quartile',
    'Maximum'
    ]]
```

```python
#printing the summary table
summary_tab
#-----------------------------------------------------------------#
```

```python
#Question 01
#b)

#only selecting the categorical features
categorical_cols = df.select_dtypes(include=['object'])
```

```python
#defining the summary statistics for the dataset and creating a summary table
categorical_summary = pd.DataFrame ({
    "Unique values Count": categorical_cols.nunique(),
    "Most Frequent Value": categorical_cols.mode().iloc[0],
    "Frequency of Most Frequent Value": categorical_cols.apply(lambda x: x.value_counts().iloc[0]),
    "Missing Values Count": categorical_cols.isnull().sum()

})
```

```python
#printing the summary table
categorical_summary
```
