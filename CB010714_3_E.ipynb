```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sample_data/Q1_Dataset.csv')
```

```python
#-------------------------------------------------------------#
#a)

# Convert categorical variables to numerical using one-hot encoding
df_encoded = pd.get_dummies(df.drop(columns=['Apartment_ID']), drop_first=True)

# Calculate correlation matrix
correlation_matrix = df_encoded.corr()

# Get the top correlations with Rental_Price (excluding itself)
top_factors = correlation_matrix['Rental_Price'].drop('Rental_Price').sort_values(key=abs, ascending=False)

# Show top 5 influencing factors
print(top_factors.head(5))
```

```python
#-------------------------------------------------------------#
#b)
# Calculate the correlation between apartment size and rental price
correlation_size_price = df['Size_in_Sqft'].corr(df['Rental_Price'])
print(f"Correlation between apartment size and rental price: {correlation_size_price:.2f}")

plt.figure(figsize=(10, 6))
sns.regplot(x='Size_in_Sqft', y='Rental_Price', data=df, scatter_kws={'alpha':0.6}, line_kws={'color': 'green'})
plt.title('Rental Price vs. Apartment Size')
plt.xlabel('Apartment Size (Sqft)')
plt.ylabel('Rental Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#-------------------------------------------------------------#
#c)
# Calculate the correlation coefficient
correlation_dist_price = df['Distance_to_City_Center'].corr(df['Rental_Price'])
print(f"Correlation between distance and rental price: {correlation_dist_price:.2f}")

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Distance_to_City_Center', y='Rental_Price', data=df, scatter_kws={'alpha':0.6}, line_kws={'color': 'green'})
plt.title('Rental Price vs. Distance to City Center')
plt.xlabel('Distance to City Center (km)')
plt.ylabel('Rental Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
