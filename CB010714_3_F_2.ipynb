```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sample_data/Q1_Dataset.csv')
df = df.drop_duplicates()
duplicate_rows = df[df.duplicated()]
print("Duplicate rows found:", len(duplicate_rows))

df['Furnished'] = df['Furnished'].str.strip().str.title()

# 1. Group by Furnished status and calculate average distance to city center
avg_distance = df.groupby('Furnished')['Distance_to_City_Center'].mean()
print("Average distance to city center by furnishing status:")
print(avg_distance)

# 2. Count of listings per Furnished status
print("\nNumber of listings by furnishing status:")
print(df['Furnished'].value_counts())

# 3. Visualize with boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Furnished', y='Distance_to_City_Center', data=df)
plt.title('Distance to City Center by Furnishing Status')
plt.xlabel('Furnished')
plt.ylabel('Distance to City Center (km)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
